```python
import pandas as pd
import numpy as np
import tensorflow as tf
```

```python
data=pd.read_csv('cardio_train.csv')
data.head()
```

```python
data.isnull().sum()
```

```python
x=data.iloc[:,1:-1]
y=data.iloc[:,-1]
```

```python
from sklearn.preprocessing import MinMaxScaler
scler=MinMaxScaler()
x=scler.fit_transform(data)
x
```

```python
y = pd.get_dummies(y)
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=3234)
print("train: shape",x_train.shape,y_train.shape)
print("test shape:",x_test.shape,y_test.shape,)
```

```python
np.shape(x)
```

```python
model=tf.keras.models.Sequential([
        tf.keras.layers.Dense(13, activation='relu',input_shape=(13,)),
    tf.keras.layers.Dense(30, activation='relu'),
    tf.keras.layers.Dense(2, activation='softmax')

])
```

```python
model.compile(optimizer=tf.keras.optimizers.Adam(0.001),loss=tf.keras.losses.CategoricalCrossentropy(from_logits=False),metrics=[tf.keras.metrics.CategoricalAccuracy()])
```

```python
model.fit(x_train,y_train,epochs=10,validation_data=(x_test,y_test),batch_size=800,verbose=3)
```

```python
from sklearn.ensemble import RandomForestClassifier
yr=data.iloc[:,-1]
```

```python
np.unique(yr,return_counts=True)
```

```python
m=RandomForestClassifier(n_estimators=200)
xr_train,xr_test,yr_train,yr_test=train_test_split(x,yr,random_state=3234)
```

```python
m.fit(xr_train,yr_train)
```

```python
from sklearn.metrics import classification_report
y_pred=m.predict(xr_test)
print(classification_report(yr_test,y_pred))
```

```python
model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(13, activation='relu', input_shape=(13,)),
    tf.keras.layers.Dense(30, activation='relu'),
     tf.keras.layers.Dense(60, activation='relu'),
     tf.keras.layers.Dense(25, activation='relu'),
    tf.keras.layers.Dense(2, activation='softmax')

])
```

```python
model.compile(optimizer=tf.keras.optimizers.Adam(0.001),loss=tf.keras.losses.CategoricalCrossentropy(from_logits=False),metrics=[tf.keras.metrics.CategoricalAccuracy()])
```

```python
model.fit(x_train,y_train,epochs=500,validation_data=(x_test,y_test),batch_size=300)
```
