```python
import tensorflow as tf
import tensorflow_datasets
```

```python
(train,test),info=tensorflow_datasets.load('mnist',split=['train','test'],as_supervised=True,with_info=True)
```

```python
train
```

```python
def normalize(image,label):
    return tf.cast(image,tf.float32)/256,label
train=train.map(normalize)
train
```

```python
train=train.cache()
```

```python
info.splits['train']
```

```python
train=train.shuffle(info.splits['train'].num_examples)
```

```python
train=train.batch(256)
```

```python
train=train.prefetch(tf.data.experimental.AUTOTUNE)
```

```python
train
```

```python
test=test.map(normalize)
```

```python
test=test.batch(256)
```

```python
test=test.cache()
test=test.prefetch(tf.data.experimental.AUTOTUNE)
```

```python
model=tf.keras.models.Sequential([
    tf.keras.layers.Flatten(input_shape=(28,28)),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(10, activation='softmax')

])
```

```python
model.compile(optimizer=tf.keras.optimizers.Adam(0.001),loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),metrics=[tf.keras.metrics.SparseTopKCategoricalAccuracy()])
```

```python
model.fit(train,epochs=10,validation_data=test)
```
